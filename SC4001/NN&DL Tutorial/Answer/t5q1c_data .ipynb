# Tutorial 5 Question 1c_data

```python
import numpy as np
import pylab as plt
```

```python
# parameters
no_features = 2

# for reproducibility
seed = 10
np.random.seed(seed)
```

```python
# generate training data
X = np.zeros((10*10, no_features))
no_data = 0
for i in np.arange(-1.0, 1.001, 2.0/9.0):
    for j in np.arange(-1.0, 1.001, 2.0/9.0):
        X[no_data] = [i, j]
        no_data += 1

Y = np.zeros((no_data, 1))
Y[:,0] = np.sin(np.pi*X[:,0])*np.cos(2*np.pi*X[:,1])
```

```python
idx = np.arange(no_data)
np.random.shuffle(idx)
X, Y = X[idx], Y[idx]

Xtest,  Ytest  = X[:34],   Y[:34]
Xtrain, Ytrain = X[34:77], Y[34:77]
Xvalid, Yvalid = X[77:],   Y[77:]
```

```python
# plot training, validation and test inputs
plt.figure(1)
plt.plot(Xtrain[:,0], Xtrain[:,1], 'rx', label='train')
plt.plot(Xvalid[:,0], Xvalid[:,1], 'bx', label='validate')
plt.plot(Xtest[:,0],  Xtest[:,1],  'gx', label='test')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Training, validation and test inputs')
plt.legend(loc='lower left')
plt.show()
```
