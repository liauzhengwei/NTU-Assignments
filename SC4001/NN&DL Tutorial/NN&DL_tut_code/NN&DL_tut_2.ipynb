```python
# c)
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # for 3D plotting

# Parameters
n_samples = 100
n_features = 3 # 3D input
n_classes = 2
random_state = 42

# Generate data
X,y = make_blobs(n_samples=n_samples,
                 n_features=n_features,
                 centers=n_classes,
                 cluster_std=5.0,
                 random_state=random_state)

# Create a 3D figure
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Plot the Class 0 samples
ax.scatter(X[y==0, 0], X[y==0, 1], X[y==0, 2], color='blue', label='Class 0')

# Plot the Class 1 samples
ax.scatter(X[y==1, 0], X[y==1, 1], X[y==1, 2], color='red', label='Class 1')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of Class 0 and Class 1')

# Legend
ax.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Convert to torch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1,1) # shape (100,1)

# Define the logistic neuron
class LogisticNeuron(nn.Module):
  def __init__(self, input_dim):
    super().__init__()
    self.linear = nn.Linear(input_dim, 1) # 3 input features -> 1 output
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    out = self.linear(x)
    out = self.sigmoid(out)
    return out

model = LogisticNeuron(input_dim=3)

# Loss and Optimizer
criterion = nn.BCELoss() # Binary Cross Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=0.01) # Stochastic Gradient Descent

epochs = 1000

for epoch in range(epochs):
  # Forward pass
  outputs = model(X_tensor)
  loss = criterion(outputs, y_tensor)

  # Backward pass and optimization
  optimizer.zero_grad() # reset gradients
  loss.backward()       # compute gradients
  optimizer.step()      # update weights

  if (epoch+1) % 50 == 0:
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

with torch.no_grad():
  y_pred = model(X_tensor)
  predicted_labels = (y_pred > 0.5).int().view(-1) # shape(100,)

# True labels as integers
y_true = y_tensor.view(-1).int()

# Classification error
num_incorrect = (predicted_labels != y_true).sum().item()
classification_error = num_incorrect / len(y_true)

print(f"\nClassification Error: {classification_error*100:.2f}%")
```

```python
# For 3D, a plane is the decision boundary.
# When the sigmoid of the neuron is 0.5, the sigmoid input is 0
# w1x1 + w2x2 + w3x3 + b = 0
# Need to solve for x3

# Extract weights and bias
w = model.linear.weight.detach().numpy()[0]
b = model.linear.bias.detach().numpy()[0]
## .linear() is the linear layer of the model
## .detach() removes the tensor from the computation graph so gradients are not tracked for it
## .numpy() makes it a numpy array
## [0] selects the first and only row to get 1D array of weights

# Create a grid for x1 and x2
x1_range = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 30)
x2_range = np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, 30)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Compute x3 values on the decision boundary plane
x3_grid = -(w[0] * x1_grid + w[1] * x2_grid + b) / w[2]

# Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Plot Gaussian points
# Plot the Class 0 samples
ax.scatter(X[y==0, 0], X[y==0, 1], X[y==0, 2], color='blue', label='Class 0')
# Plot the Class 1 samples
ax.scatter(X[y==1, 0], X[y==1, 1], X[y==1, 2], color='red', label='Class 1')

# Plot the decision boundary plane
ax.plot_surface(x1_grid, x2_grid, x3_grid, color='green', alpha=0.8,edgecolor="k", label='Decision Boundary')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of Class 0 and Class 1 with Decision Boundary')
ax.legend()
plt.show()
```

```python
# d)
import numpy as np
import matplotlib.pyplot as plt

# Function
def f(x,y):
  return 1.5 + 3.3*x -2.5*y + 1.2*x*y

# Input bounds
x_min, x_max = 0.0, 10.0
y_min, y_max = -10.0, 1.0

# Sample data
n_samples = 100
rng = np.random.default_rng(42)
x = rng.uniform(low=x_min, high=x_max, size=n_samples)
y = rng.uniform(low=y_min, high=y_max, size=n_samples)
X = np.column_stack((x,y))

# Labels in {-1, +1}
labels = np.where(f(x,y) >= 0, 1, -1)

# Perceptron training
w = np.zeros(2)
b = 0.0
eta = 0.01
epochs = 1000

for epoch in range(epochs):
  errors = 0
  for xi, target in zip(X, labels):
    activation = np.dot(w, xi) + b
    y_prediction = 1 if activation >=0 else -1
    if y_prediction != target:
      errors += 1
      w += eta * target * xi
      b += eta * target
  print(f"Epoch {epoch+1}, errors={errors}")
  if errors == 0:
    break

print("Learned weights:", w)
print("Learned bias:", b)

# Classification error
y_preds = np.sign(X.dot(w) + b)
classification_error = np.mean(y_preds != labels)
print(f"Classification Error: {classification_error*100:.2f}%")
```

```python
# Plot results
plt.figure(figsize=(8,6))
plt.scatter(X[labels==1,0], X[labels==1,1], color='blue', label='Class +1')
plt.scatter(X[labels==-1,0], X[labels==-1,1], color='red', label='Class -1')

# Decision boundary line
if abs(w[1]) > 1e-12:
  x_vals = np.linspace(x_min-1, x_max+1, 200)
  y_vals = -(w[0]/w[1])*x_vals - b/w[1]
  plt.plot(x_vals, y_vals, 'k-', linewidth=2, label='Decision Boundary')
else:
  plt.axvline(x=-b/w[0], color='k', linewidth=2, label='Decision Boundary')

plt.xlim(x_min-1, x_max+1)
plt.ylim(y_min-1, y_max+1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Perceptron Classification Results')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Linear Neuron
def target_function(x,y):
  return 1.5 + 3.3*x -2.5*y + 1.2*x*y

# Sample 100 random points
n_samples = 100
rng = np.random.default_rng(42)
x = rng.uniform(0, 10, n_samples)
y = rng.uniform(-10, 1, n_samples)
X = np.column_stack((x,y))

# Labels
labels = np.where(target_function(x,y) >= 0, 1, -1)

# Linear neuron training (gradient descent on MSE)
w = np.zeros(2)
b = 0.0
eta = 0.01
epochs = 1000

for epoch in range(epochs):
  y_pred = np.dot(X, w) + b
  error = labels - y_pred

  w += eta * np.dot(X.T, error)/n_samples
  b += eta * np.mean(error)

  mse = np.mean(error**2)
  print(f"Epoch {epoch+1}, MSE={mse}")

print("Learned weights:", w)
print("Learned bias:", b)

y_class = np.where(y_pred >=0, 1, -1)
classification_error = np.mean(y_class != labels)
print(f"Classification Error: {classification_error*100:.2f}%")

plt.figure(figsize=(8,6))
plt.scatter(x[labels==1], y[labels==1], color='blue', label='Class +1')
plt.scatter(x[labels==-1], y[labels==-1], color='red', label='Class -1')

# Decision Boundary
x_vals = np.linspace(x_min, x_max, 200)
if w[1] != 0:
  y_vals = -(w[0]/w[1])*x_vals - b/w[1]
  plt.plot(x_vals, y_vals, 'k-', linewidth=2, label='Decision Boundary')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Neuron Classification Results')
plt.legend()
plt.grid(True)
plt.show()
```
