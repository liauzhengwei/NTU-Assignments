```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#b

df=pd.read_csv("Proj2\\stuff\\bVaryingVertices.csv")
df['Average Key Comparisons'] = df['Average Key Comparisons'].astype(int)
df['Number of Vertices']=df['Number of Vertices'].astype(int)

vertices = np.array(range(10,1000))
tkeycomp = ((3/8)*(vertices*vertices-vertices)) * np.log2(vertices) #ElogV

fig, ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
plt.plot(df["Number of Vertices"].values.tolist(), df["Average Key Comparisons"].values.tolist(), label = 'Empirical Data')
plt.plot(vertices, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Varying Vertices')
plt.xlabel('Number of Vertices')
plt.ylabel('Average Key Comparisons')


plt.legend()
plt.show()
```

```python
df=pd.read_csv("Proj2\\stuff\\bVaryingEdgesKeyComp.csv")
df['Average Key Comparisons'] = df['Average Key Comparisons'].astype(int)
df['Number of Edges']=df['Number of Edges'].astype(int)

edges = np.array(range(100,9900))
tkeycomp = edges * np.log2(100) #ElogV

fig, ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
plt.plot(df["Number of Edges"].values.tolist(), df["Average Key Comparisons"].values.tolist(), label = 'Empirical Data')
plt.plot(edges, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Varying Edges')
plt.xlabel('Number of Edges')
plt.ylabel('Average Key Comparisons')


plt.legend()
plt.show()
```
