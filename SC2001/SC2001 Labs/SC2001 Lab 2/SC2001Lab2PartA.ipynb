```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#unidirectional
df = pd.read_csv("C:\\Users\\ZW\\Desktop\\SC2001Lab2\\AllCSV\\adj matrix\\comparisons\\1direction\\aVaryVFixedEuniDirectionalComparisons.csv")
df['Average Number of Comparisons Done'] = df['Average Number of Comparisons Done'].astype(int)

tsize = np.array(range(1,1000))
tkeycomp = tsize*tsize

fig,ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
plt.plot(df["Number of Vertices"].values.tolist(), df["Average Number of Comparisons Done"].values.tolist(), label = 'Empirical Data')
plt.plot(tsize, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Number of Vertices')
plt.xlabel('Number of Vertices')
plt.ylabel('Average Key Comparisons Done')
plt.xticks(range(min(df["Number of Vertices"].values.tolist()), max(df["Number of Vertices"].values.tolist()) + 1,100),rotation='vertical')  # Change the step value as needed
plt.yticks(range(0, 4500000,500000 ))


plt.legend()
plt.show()
```

```python
#unidirectional Vary E fixed V
df = pd.read_csv("C:\\Users\\ZW\\Desktop\\SC2001Lab2\\AllCSV\\adj matrix\\comparisons\\aFixedVVaryEcomparisons.csv")
df['Average Number of Comparisons'] = df['Average Number of Comparisons'].astype(int)

tsize = np.array(range(1,10000))
tkeycomp = np.full(9999,10000)

fig,ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
plt.plot(df["Number of Edges for 100 Vertices"].values.tolist(), df["Average Number of Comparisons"].values.tolist(), label = 'Empirical Data')
plt.plot(tsize, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Number of Edges')
plt.xlabel('Number of Edges')
plt.ylabel('Average Key Comparisons Done')
plt.xticks(range(0, max(df["Number of Edges for 100 Vertices"].values.tolist()) + 1,500),rotation='vertical')  # Change the step value as needed
plt.yticks(range(0, 21000,1000 ))


plt.legend()
plt.show()
```
