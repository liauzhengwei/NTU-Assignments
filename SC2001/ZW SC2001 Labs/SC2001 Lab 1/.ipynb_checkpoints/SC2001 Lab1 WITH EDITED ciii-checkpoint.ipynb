```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#c(i)

#100 plot points from 1000 to 10,000,000 at intervals of 99990, each value is obtained by creating 100 different arrays and 
#taking the average result of the key comparisons. 

df=pd.read_csv("Proj1\ci\ci.csv")
df['Average Key Comparisons'] = df['Average Key Comparisons'].astype(int)

tsize = np.array(range(1000,10000000))
tkeycomp = tsize*5 + tsize*(np.log2(tsize/5))

fig, ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
plt.plot(df["Array Size"].values.tolist(), df["Average Key Comparisons"].values.tolist(), label = 'Empirical Data')
plt.plot(tsize, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Array Size')
plt.xlabel('Array Size')
plt.ylabel('Average Key Comparisons')
plt.xticks(range(min(df["Array Size"].values.tolist()), max(df["Array Size"].values.tolist()) + 1, 666600),rotation='vertical')  # Change the step value as needed
plt.yticks(range(0, 350000000,50000000 ))


plt.legend()
plt.show()
```

```python
#c(ii)
#100 plot points from 1000 to 10,000,000 at intervals of 99990, each value is obtained by creating 100 different arrays and 
#taking the average result of the key comparisons. Similar to above, but here we vary S instead of Array Size.

df=pd.read_csv("Proj1\cii\cii.csv")
df['Average Key Comparisons'] = df['Average Key Comparisons'].astype(int)

s = np.array(range(1,250))
tkeycomp = 100000*s + 100000*(np.log2(100000/s))

fig, ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')

plt.plot(df["Value of S"].values.tolist(), df["Average Key Comparisons"].values.tolist(), label = 'Empirical Data')
plt.plot(s, tkeycomp, label = 'Theoretical Data')

plt.title('Plot of Key Comparisons against Value of S')
plt.xlabel('Value of S')
plt.ylabel('Average Key Comparisons')
plt.xticks(range(min(df["Value of S"].values.tolist()), 260,10),rotation='vertical')  # Change the step value as needed

plt.legend()
plt.show()
```

```python
#c(iii)

#We obtain data on 25 different array sizes of interval 399960. 
#For each array size, we generate 21 random arrays of that array size, and calculate the time taken for every value of S,
#from 0-250.
#Then we calculate the average time taken for the 21 iterations for each value of S for that array size. 
#The overall trend seems to suggest that for small values of S,(S<10?) the time taken is significantly longer, and 
#the optimal value of S lies rather consistantly within the 35-55 range.

start_i = 1000
end_i = 10000000
interval_i = 399960
# Calculate the number of subplots needed
num_subplots = (end_i - start_i) // interval_i + 1

# Create a figure and specify the number of rows and columns for subplots
num_rows = 9  # Change this as needed
num_cols = 3 # Change this as needed
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, 50))

# Iterate over 'i' values and create subplots
for i, ax in zip(range(start_i, end_i + 1, interval_i), axs.flatten()):
    csvName = f"Proj1/ciii/{i}.csv"
    df = pd.read_csv(csvName)
    ax.plot(df["S value"].values.tolist(), df["Time taken"].values.tolist())
    ax.set_title(f'Plot of Time Taken against S Value (Array Size={i})')
    ax.set_xlabel('S value')
    ax.set_ylabel('Time taken')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()




```

```python
#c(iii)

#We repeat for key comparisons, to show that lower key comp doesnt translate to faster speed.

start_i = 1000
end_i = 10000000
interval_i = 399960
# Calculate the number of subplots needed
num_subplots = (end_i - start_i) // interval_i + 1

# Create a figure and specify the number of rows and columns for subplots
num_rows = 9  # Change this as needed
num_cols = 3 # Change this as needed
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, 50))


# Iterate over 'i' values and create subplots
for i, ax in zip(range(start_i, end_i + 1, interval_i), axs.flatten()):
    csvName = f"Proj1/ciii keycomp/{i}.csv"
    df = pd.read_csv(csvName)
    ax.plot(df["S value"].values.tolist(), df["Number of Comparisons"].values.tolist())
    ax.set_title(f'Plot of Number of Key Comparisons against S Value (Array Size={i})')
    ax.set_xlabel('S value')
    ax.set_ylabel('Key Comparisons')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
#We create 100 random arrays of size 10,000,000. We conduct both mergesort and hybridsort on each array.
#We collect the time taken, and key comparisons for both mergesort and hybridsort of each array
#Then we take the average result of key comparisons and time taken from the 100 iterations.

df=pd.read_csv("Proj1\d\d.csv")

#Plot Avg HBS time against Avg MS Time
x_axis=["Hybrid Sort","Merge Sort"]
y_axis=[df.loc[0]['Avg HBS Time'],df.loc[0]['Avg MS Time'] ]


plt.bar(x_axis, y_axis)
plt.xlabel("Sorting Algorithm")
plt.ylabel("Time Taken")
plt.title("Plot of Time Taken against Sorting Algorithm")
plt.ylim(0, 2500)
plt.show()
```
