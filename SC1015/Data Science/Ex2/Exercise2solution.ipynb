```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
print(train.dtypes)
train.info()
```

```python
inttrain = train.select_dtypes('int64')
inttrain
```

Actual Numeric Values: 
Id, LotArea, YrSold, YearBuilt, YearRemodAdd, BsmtFinSF1,BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, 1stFlrSF,2ndFlrSF, LowQualFinSF,GrLivArea, BsmtFullBath,BsmtHalfBath, FullBath,HalfBath,BedroomAbvGr, KitchenAbvGr, TotRmsAbvGrd, Fireplaces, GarageCars, GarageArea, WoodDeckSF, OpenPorchSF, EnclosedPorch,3SsnPorch, ScreenPorch, PoolArea, MiscVal, SalePrice


Categorical Variables: 
MSSubClass, MoSold, OverallQual,OverallCond,

```python
inttrainclean = inttrain.copy()
```

```python
inttrainclean1 = inttrainclean.loc[:,inttrainclean.columns!='MoSold']
inttrainclean2 = inttrainclean1.loc[:,inttrainclean1.columns!='MSSubClass']
inttrainclean3 = inttrainclean2.loc[:,inttrainclean2.columns!='OverallQual']
inttrainclean4 = inttrainclean3.loc[:,inttrainclean3.columns!='OverallCond']

print(inttrainclean4)
```

```python
SP= pd.DataFrame(inttrainclean4['SalePrice'])
SP.describe()
```

```python
f,axes = plt.subplots(1,1, figsize = (12,6))
sb.boxplot(data = SP, orient = "h")

f=plt.figure(figsize=(12,6))
sb.histplot(data = SP)

f = plt.figure(figsize = (12,6))
sb.kdeplot(data = SP)
```

```python
LA = pd.DataFrame(inttrainclean4['LotArea'])
LA.describe()
```

```python
f,axes = plt.subplots(1,1, figsize = (12,6))
sb.boxplot(data = LA, orient = "h")

f=plt.figure(figsize=(12,6))
sb.histplot(data = LA)

f = plt.figure(figsize = (12,6))
sb.kdeplot(data = LA)
```

```python
jointSPLA = pd.concat([SP,LA],axis =1)
jointSPLA
```

```python
sb.jointplot(data = jointSPLA,x = "LotArea",y="SalePrice",height = 12)
```
